```python
#players in each country

player_country = df[['sofifa_id', 'nationality']]
player_country.head(5)
```

```python
country_list = player_country['nationality'].unique()
country_list.sort()
country_list
```

```python
#there are 162 unique countries in the dataset
len(country_list)
```

```python
#number of players in each country
count_by_country = player_country.groupby(by='nationality').count().rename(columns={'sofifa_id':'player_count'})
count_by_country.head(5)
```

```python
#Top 10 player counts
count_by_country = count_by_country.sort_values(by='player_count', ascending=False)
count_by_country.head(10)
```

```python
#top 5 countries with the most players bar plot
top_count_plot = count_by_country.sort_values(by='player_count', ascending=False).head(5).reset_index()
top_count_plot.plot.bar(x='nationality', y='player_count', color="green")
plt.show()
```

```python
df['overall'].unique()
```

```python
#top 5 players by overall score
top_players = df[['sofifa_id', 'short_name', 'wage_eur', 'overall']].sort_values(by='overall', ascending=False)
top_players.head(5)
```

## Top 5 Salaries

```python
#top 5 player salaries
top_salaries = top_players[['sofifa_id', 'short_name', 'wage_eur']].sort_values(by='wage_eur', ascending=False)
top_salaries.head(5)
```

```python
#top 5 player salaries bar plot
top_salaries_plot = top_salaries.head(5).reset_index()

# function to add formatting for labels on bars
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i]//2,y[i], ha = 'center')
        
x = top_salaries_plot['short_name']
y = top_salaries_plot['wage_eur']

plt.figure(figsize=(10,6))
bar_chart = plt.bar(x, y)
addlabels(x, y)

# giving title to the plot
plt.title("Top 5 Player Salaries")
    
# giving X and Y labels
plt.xlabel("Short Name")
plt.ylabel("Salary (EUR)")

plt.show()
```

## Top 5 Shooters & Defenders

```python
#Top 5 best shooters with short names

#To ensure i'm only getting records of those with short names as instructed, I use the dropna with short_name as the subset to drop the na records
top5_shooters = df[["sofifa_id","short_name","shooting"]].dropna(subset=['short_name']).sort_values(by="shooting", ascending=False).head(5)
top5_shooters
```

```python
#Top 5 best defender player records
top5_defenders = df[["sofifa_id","short_name","defending","nationality","club"]].sort_values(by="defending", ascending=False).head(5)
top5_defenders
```

## Real Madrid Analysis

```python
#Making a dataframe for Real Madrid team top 5 player queries

#Based on instruction, we will be doing analysis on shooting and defending stats, so I will dropna based on those subset columns since there are missing statistics
top5_realmadrid_df = df.loc[df["club"]=="Real Madrid"].sort_values(by="overall", ascending=False).dropna(subset=["shooting", "defending"]).head(5)
top5_realmadrid_df.head(5)
```

```python
top5_realmadrid_wages = top5_realmadrid_df[["sofifa_id", "short_name", "wage_eur"]].sort_values(by="wage_eur", ascending=False).head(5)
top5_realmadrid_wages 
```

```python
top5_realmadrid_shooters = top5_realmadrid_df[["sofifa_id", "short_name", "shooting"]].sort_values(by="shooting", ascending=False).dropna().head(5)
top5_realmadrid_shooters 
```

```python
top5_realmadrid_defenders = top5_realmadrid_df[["sofifa_id", "short_name", "defending"]].sort_values(by="defending", ascending=False).head(5)
top5_realmadrid_defenders 
```
