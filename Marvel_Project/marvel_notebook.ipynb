```python
#getting distinct counts of each Alignment type
df.value_counts(subset="Alignment")
```

```python
#making a dataframe for only the good alignment characters

good_chars = df.loc[df["Alignment"]=="good"]
good_chars.head(5)
```

```python
# top speeds with good alignment

top_speeds_good = df.loc[df["Alignment"]=="good"].sort_values(by='Speed', ascending=False).head(5)
top_speeds_good
```

```python
# superheros with good alignment & Max power

top_power_good = df.loc[df["Alignment"]=="good"].sort_values(by='Power', ascending=False)
top_power_good.head(5)
```

```python
# total count for 100 power and good alignment

total_top_power_good = top_power_good.loc[top_power_good["Power"]==100].value_counts(subset="Power")
total_top_power_good
```

```python
#power 100 good alignment shape

power_100_good = top_power_good.loc[top_power_good["Power"]==100]
power_100_good.shape
```

```python
#power 100 good alignment records

power_100_good
```

```python
power_100_good.head(5)
```

```python
#top 5 player salaries bar plot
power_100_good_plot = power_100_good[['Name','Total']].head(5).sort_values(by='Total').reset_index()

# function to add formatting for labels on bars
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i]//2,y[i], ha = 'center')
        
x = power_100_good_plot['Name']
y = power_100_good_plot['Total']

plt.figure(figsize=(10,6))
bar_chart = plt.bar(x, y)
addlabels(x, y)

# giving title to the plot
plt.title("Top 5 Good/Max Power Superheros")
    
# giving X and Y labels
plt.xlabel("Name")
plt.ylabel("Total")

plt.show()
```

```python
# making a dataframe for only the bad alignment characters

bad_chars = df.loc[df["Alignment"]=='bad']
bad_chars.head(5)
```

```python
# plotting a histogram of the distribution of speeds amongst good alignments

plt.figure(figsize=(10.5, 5))
plt.hist(good_chars['Speed'], bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of Speed")
plt.xlabel("Speed")
plt.ylabel("Frequency")
plt.show()
```
